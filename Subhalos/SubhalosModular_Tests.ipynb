```python
## First, check my inverse transform
# If I feed it a distribution, it should return a sample that follows the distribution

test_vals = np.linspace(0,1,100)
def pdf(x):
    return x

test_sampler = InverseTransform.InverseTransform(pdf, test_vals, nsamples=100000);
plt.hist(test_sampler.sample(), bins=20)
plt.plot(test_vals, 500000*pdf(test_vals)/np.sum(pdf(test_vals)))
## Looks good!
```
